```python
import numpy as np
from reward import KL_divergence_reward
from utils import parametrize_qiskit_circuit
from concurrent.futures import ThreadPoolExecutor

# Calculate and print rewards for each circuit
for j, circuit in enumerate(circuits["results"]):
    qasm_str = circuit["generated_circuit"]
    random_circuits = []
    try:
        qiskit_circuit = loads(qasm_str)
        parametrized_circuit, param_map = parametrize_qiskit_circuit(qiskit_circuit)
        
        def generate_random_circuit(_):
            param_map_random = {param: np.random.uniform(-np.pi, np.pi) for param in param_map.values()}
            new_parametrized_circuit = parametrized_circuit.assign_parameters(param_map_random)
            return dumps(new_parametrized_circuit)
        
        with ThreadPoolExecutor() as executor:
            random_circuits = list(executor.map(generate_random_circuit, range(100)))

    except Exception as e:
        print(f"Error loading circuit {j}: {e}")
        continue
    ground_truth_index = circuit["sample_index"]
    test_item = test_dict[ground_truth_index]
    try:
        reward = KL_divergence_reward(qasm_str, test_item)

        def calculate_random_reward(rc):
            return KL_divergence_reward(rc, test_item)
        
        with ThreadPoolExecutor() as executor:
            random_rewards = list(executor.map(calculate_random_reward, random_circuits))
        
        avarage_random_reward = np.mean(random_rewards)
    except Exception as e:
        print(f"Error calculating reward for circuit {j}: {e}")
        continue
    print(f"Circuit {j} reward: {reward}", f"Average random reward: {avarage_random_reward}")
    rewards[ground_truth_index] = {"KL_divergence_reward": reward, "random_rewards": avarage_random_reward}
```

```python
with open("circuit_rewards.json", "w") as f:
    json.dump(rewards, f, indent=4)
```

```python
# Compute avarage reward and avarage random reward and compare difference
with open("circuit_rewards.json") as f:
    rewards = json.load(f)
reward_values = [v["KL_divergence_reward"] for v in rewards.values() if v["KL_divergence_reward"] != "inf"]
random_reward_values = [v["random_rewards"] for v in rewards.values() if v["random_rewards"] != "inf"]
average_reward = np.mean(reward_values)
average_random_reward = np.mean(random_reward_values)
print(f"Average reward: {average_reward}")
print(f"Average random reward: {average_random_reward}")
print("Improvement over random:", (1 - average_reward / average_random_reward) * 100, "%")
```
