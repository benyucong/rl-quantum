````python
from datasets import load_dataset
from reward import expectation_value_reward, KL_divergence_reward, syntax_reward, optimization_reward_pennylane, optimization_reward_qiskit

ds = load_dataset("valterUo/graph-data-quantum-updated")

for i in range(10):
    llm_qasm_circuit_string = ds["train"][i]["circuit_with_params"]
    ground_truth = ds["train"][i]
    print("Syntax reward: ", syntax_reward(llm_qasm_circuit_string))
    print("Entropy reward: ", KL_divergence_reward(llm_qasm_circuit_string, ground_truth))
    print("Expecation reward: ", expectation_value_reward(llm_qasm_circuit_string, ground_truth))
    print("Training reward: ", optimization_reward_pennylane(llm_qasm_circuit_string, ground_truth))
    print("Training reward: ", optimization_reward_qiskit(llm_qasm_circuit_string, ground_truth))
````
