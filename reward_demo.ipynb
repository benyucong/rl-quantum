````python
from datasets import load_dataset

ds = load_dataset("linuzj/graph-data-quantum")
````

````python
# To demonstrate the reward function, let's take a circit from the dataset
llm_qasm_circuit_string = ds["train"][0]["circuit_with_params"]
ground_truth = ds["train"][0]
````

````python
from reward import syntax_reward


print(syntax_reward(llm_qasm_circuit_string))  # Should print 1.0 for valid QASM syntax
````

````python
from reward import KL_divergence_reward


print(KL_divergence_reward(llm_qasm_circuit_string, ground_truth))
````

````python
from reward import expectation_value_reward


print(expectation_value_reward(llm_qasm_circuit_string, ground_truth))
````
