```python
import json

hubo_problem_indices = []
with open("quantum_circuits_output_20250911_144036_rl_quantum_4b.json", "r") as f:
    data = json.load(f)

for problem in data["results"]:
    if problem["dataset_metrics"]["problem_type"] == "edge_cover":
        hubo_problem_indices.append(problem["sample_index"])

print("Hubo problem indices:", len(hubo_problem_indices))

with open('evaluation_results.json') as f:
    results = json.load(f)

results_by_type = {}
for result in results:
    for key in results[result]:
        if key not in results_by_type:
            results_by_type[key] = []

        if results[result][key] is not None and results[result]["llm_js_divergence_reward"] is not None and int(result) in hubo_problem_indices:
            results_by_type[key].append(results[result][key])

for key in results_by_type:
    avg = sum(results_by_type[key]) / len(results_by_type[key])
    print(f"{key}: {avg}", "Number of samples:", len(results_by_type[key]))
```
